## Reconfigure a Spatial Grid, adding values for Row and Column Indices

This code reads a Geopackage, and adds new columns to give index values to features in the grid.

```python
import geopandas as gpd
import pandas as pds
```

```python
# read in Geopackage file to work with
df = gpd.read_file('/media/eubtube/Seagate Backup Plus Drive/Congo_Tifs/vector/colline_grid_joined.gpkg')
```

```python
# Get unique y coordinates for columns in grid
cols = df['left'].unique()
cols
```

```python
# Create keys for dictionary in next step (will become the column identifiers)
keys = list(range(0,34))
keys
```

```python
# create the dictionary 
new_dict = dict(zip(keys, cols))
new_dict
#flip the keys and values in a dict
new_dict2 = dict((y,x) for x,y in new_dict.items())
new_dict2
```

```python
# Create the 'Cols' Column by applying the key value from dictionary where the x coordinate exists
df["cols"] = df["left"].apply(lambda x: new_dict2.get(x))
df 
```

```python
# generate row id based on column id
# 32 rows per column
df['grid_id1'] = df['id']-(32 * (df['cols']-1))
df
```

```python
# Drops a column
# df.drop(['grid_id'], axis=1, inplace=True)
df
```

```python
# as a tuple
df['grid_id'] = zip(df['grid_id1'], df['cols'])
df
```

```python
# as a list
df['grid_id'] = df[['grid_id1', 'cols']].values.tolist()
```

```python
# Creates a column with string version of 'grid_id1' and 'cols'
df['Grid_Index'] = df.agg(lambda x: f"{x['grid_id1']}, {x['cols']}", axis=1)
```

```python
# Drop Unneeded columns
df.drop('grid_id', axis=1, inplace=True)
```

```python
#rename columns
# df.rename({'grid_id1':'rows'}, axis=1, inplace=True)

df.rename({'cols':'col'}, axis=1, inplace=True)
```

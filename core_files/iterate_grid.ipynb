```python
# Make Row and Col values int

df['row'] = df['row'].astype(int)
df['col'] = df['col'].astype(int)
df
```

```python
# Add "grouprow" column for filter iteration
df['grouprow'] = df['row'].apply(lambda x: np.floor((x-1)/3+1).astype(int))
df
```

```python
# Add "groupcol" column for filter iteration
df['groupcol'] = df['col'].apply(lambda x: np.floor((x-1)/3+1).astype(int))
df
```

```python
# Pass a range of rows and columns, 
# spit the tiles associated with quadkeys at each position into an aoi file.
# How to only accept a range or coerce into a range?, exclusions?

def aoi(rowrange, colrange): 
    df_clipped = df[(df['row'] == rowrange[0]) & (df['col'] == colrange[0])]
    for x in rowrange:
        for y in colrange:
            df_clipped = df_clipped.append(df[(df['row'] == x) & (df['col'] == y)])
    
    df_clipped.drop(df_clipped.index[0],inplace=True)
    
    return df_clipped

aoi(range(7,10), range(9,14))
```

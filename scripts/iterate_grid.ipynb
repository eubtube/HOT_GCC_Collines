```python
# Make"grouprow" and "groupcol" for filter iteration

def make_group_cols(df,nrow,ncol):
    # A function to make new 
    df['grouprow'] = df['row'].apply(lambda x: np.floor((x-1)/nrow+1).astype(int))
    df['groupcol'] = df['col'].apply(lambda x: np.floor((x-1)/ncol+1).astype(int))
    return df
    
make_group_cols(df,3,3)
```

```python
# Snippet that makes a column for project name
def make_ProjName_col(df):
    df['ProjectName'] = "project" + "_" + df['groupcol'].astype('string') + "_" + df['grouprow'].astype('string')

make_ProjName_col(df)

df
```

```python
## Fancy stuff, but if we just make aois by iterating through unique identifiers (created above), we won't need to do this at all.

# Pass a range of rows and columns, 
# spit the tiles associated with quadkeys at each position into an aoi file.


def aoi(r1,r2,c1,c2):
    
    rowrange=range(r1-1,(r2+1)) #add one to the second so that range stays with input.
    colrange=range(c1-1,(c2+1)) 
    
    df_clipped = df[(df['row'] == rowrange[0]) & (df['col'] == colrange[0])]
    for x in rowrange:
        for y in colrange:
            df_clipped = df_clipped.append(df[(df['row'] == x) & (df['col'] == y)])
    
    df_filter = df_clipped[df_clipped['layer'] != 0] # Filters only layers with tiles
    #df_clipped.drop(df_clipped.index[0],inplace=True)
    
    return df_filter

aoi(3,6,3,6)
```
